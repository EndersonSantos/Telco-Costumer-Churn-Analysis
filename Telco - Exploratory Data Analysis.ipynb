```python
corr_df = pd.get_dummies(telco[['gender', 'SeniorCitizen', 'Partner', 'Dependents',
        'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'Churn']]).corr()
```

```python
telco = telco.reset_index()
```

```python
plt.figure(figsize=(10,4),dpi=200)
plt.title("Churn Rate per Tenure")
yes_churn = telco.groupby(["Churn", "tenure"]).count().transpose()["Yes"]
no_churn = telco.groupby(["Churn", "tenure"]).count().transpose()["No"]
churn_rate = (100*yes_churn / (no_churn+yes_churn)).transpose()["customerID"]
churn_rate.plot()
plt.ylabel("Churn Rate")
```

```python
def cohort(tenure):
    if tenure<13:
        return "0-12 months"
    elif tenure<25:
        return "12-24 months"
    elif tenure <49:
        return "24-48 months"
    else:
        return "Over 48 Months"
```

```python
telco["Tenure Cohort"]=telco["tenure"].apply(cohort)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=telco, x="MonthlyCharges", y="TotalCharges", hue="Tenure Cohort", alpha=0.5)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=telco, x="Tenure Cohort", hue="Churn", palette="viridis")
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.catplot(data=telco, x="Tenure Cohort",hue="Churn", kind="count", col="Contract", palette="viridis")
```
